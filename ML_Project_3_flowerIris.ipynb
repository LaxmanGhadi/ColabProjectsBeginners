```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
train_path = '/content/data/bezdekIris.data'
# train = pd.read_csv(train_path, sep = '\t', names = ['sepal length','sepal width', 'ptal_length' ',petal_width',' flower'])
train = pd.read_csv(train_path)
```

```python
new_index  = ['sepal length','sepal width', 'petal_length' ,'petal_width','flower']
train.columns = new_index
```

```python
train
```

```python
train_df  = pd.DataFrame(train)
```

```python
X  = train.drop('flower' ,axis = 1)
```

```python
X
```

```python
Y =train['flower']
```

```python
X_train,X_test, Y_train , Y_test = train_test_split(X,Y, test_size = 0.2, random_state = 2,stratify= Y )
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier()
model.fit(X_train,Y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
X_test_pred = model.predict(X_test)
test_data_accuracy  = accuracy_score(X_test_pred,Y_test)
```

```python
print(test_data_accuracy)
```
